```python
#Importing data from flowJo to do statistical analysis tests
import pandas
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats
from sklearn import preprocessing

#Looking at 0 IL-12
C4 = pandas.read_csv('export_C4_iRFP+ high.csv')
C5 = pandas.read_csv('export_C5_iRFP+ high.csv')
C6 = pandas.read_csv('export_C6_iRFP+ high.csv')
C7 = pandas.read_csv('export_C7_iRFP+ high.csv')

# stats.wilcoxon(data['FSIQ'], data['PIQ']) only tests between two sets of data, so I'd look at negTF along with each of the constructs

#H0: distributions for Tbet/TCF between control and different overexpression constructs are the same
#Ha: distributions for one of them are different-- check P values against 0.05 (or 0.001)

#The P value measures the strength of evidence against H0. The smaller the P value, the stronger the evidence against H0.


stat1 = stats.mannwhitneyu(C4['TCF1-YFP-A'], C5['TCF1-YFP-A'], alternative='two-sided')
stat2 = stats.mannwhitneyu(C4['TCF1-YFP-A'], C6['TCF1-YFP-A'], alternative='two-sided')
stat3 = stats.mannwhitneyu(C4['TCF1-YFP-A'], C7['TCF1-YFP-A'], alternative='two-sided')

stat4 = stats.mannwhitneyu(C4['Tbet-AmCyan-A'], C5['Tbet-AmCyan-A'], alternative='two-sided')
stat5 = stats.mannwhitneyu(C4['Tbet-AmCyan-A'], C6['Tbet-AmCyan-A'], alternative='two-sided')
stat6 = stats.mannwhitneyu(C4['Tbet-AmCyan-A'], C7['Tbet-AmCyan-A'], alternative='two-sided')

plt.hist(C4['TCF1-YFP-A'], bins=np.logspace(np.log10(1),np.log10(1000000), 100))
plt.xscale('log')

plt.hist(C5['TCF1-YFP-A'], bins=np.logspace(np.log10(1),np.log10(1000000), 100))
plt.xscale('log')

plt.hist(C6['TCF1-YFP-A'], bins=np.logspace(np.log10(1),np.log10(1000000), 100))
plt.xscale('log')

plt.hist(C7['TCF1-YFP-A'], bins=np.logspace(np.log10(1),np.log10(1000000), 100))
plt.xscale('log')

plt.show()

#Just as a reminder: C4: neg TF , C5: TCF, C6: Tbet 
# TCF--C4-C5: statistic=340873.5, pvalue=2.489432062558746e-53
# TCF--C4-C6: statistic=555868.0, pvalue=4.4897416766058265e-10 <---
# TCF--C4-C7: statistic=874389.0, pvalue=3.209027661243582e-31
# Tbet--C4-C5: statistic=233728.0, pvalue=0.11033261236711248 <---
# Tbet--C4-C6: statistic=881379.0, pvalue=3.2018956714185286e-46
# Tbet--C4-C7: statistic=557951.5, pvalue=1.3747589050818055e-14
```
